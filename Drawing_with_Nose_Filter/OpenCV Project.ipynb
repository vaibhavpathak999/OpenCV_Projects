# Detecting a nose 

```python
import cv2
import numpy as np

nose_cascade = cv2.CascadeClassifier(r'C:\Users\vaibh\OneDrive\Desktop\ML and CV files\opencv-master\opencv-master\data\haarcascades\haarcascade_frontalface_default.xml')
    
if nose_cascade.empty():
  raise IOError('Unable to load the nose cascade classifier xml file')

cap = cv2.VideoCapture(0)
ds_factor = 0.5

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, None, fx=ds_factor, fy=ds_factor, interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    nose_rects = nose_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in nose_rects:
        width = x+w//2
        height = y+h//2
        print(width)
        print(height)
        cv2.rectangle(frame, (width,height+5), (width+1,height+1+5), (255,0,0), 2)
        break

    cv2.imshow('Nose Detector', frame)
    

    c = cv2.waitKey(1)
    if c == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

# Making a box around the nose and drawing

```python
import cv2
import numpy as np

nose_cascade = cv2.CascadeClassifier(r'C:\Users\vaibh\OneDrive\Desktop\ML and CV files\opencv-master\opencv-master\data\haarcascades\haarcascade_frontalface_default.xml')
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')    
if nose_cascade.empty():
  raise IOError('Unable to load the nose cascade classifier xml file')

cap = cv2.VideoCapture(0)
ds_factor = 0.5

def draw_rect(img,width,height):
    cv2.rectangle(img, (width,height+5), (width+1,height+1+5), (255,255,255), 4)
    cv2.rectangle(img, (width,height+5), (width+2,height+2+5), (255,255,255), 3)

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, None, fx=ds_factor, fy=ds_factor, interpolation=cv2.INTER_AREA)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    nose_rects = nose_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in nose_rects:
        width = x+w//2
        height = y+h//2
        
        cv2.rectangle(frame, (width-10,height-15), (width+10,height+15), (255,0,0), 2)
        draw_rect(img,width,height)
        break

    cv2.imshow('Nose Detector', frame)
    cv2.imshow('Draw', img)
    

    c = cv2.waitKey(1)
    if c == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
